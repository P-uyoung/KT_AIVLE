### Gradient Tapes

fit을 수동으로 해보기 (파이토치 처럼)

```python
import numpy as np
```

```python
import pandas as pd
import tensorflow as tf

path = "https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/boston.csv"
df = pd.read_csv(path)

x_train = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis',
              'rad', 'tax','ptratio', 'b', 'lstat']]
y_train = df[['medv']]
print(x_train.shape, y_train.shape)

X = tf.keras.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.Model(X, Y)
model.compile(loss='mse')

model.summary()
```

```python
loss = tf.keras.losses.MeanSquaredError()
optim = tf.keras.optimizers.Adam(learning_rate=0.001)

for e in range(1000):
  with tf.GradientTape() as tape:
    pred = model(np.array(x_train), training=True)
    cost = loss(np.array(y_train), pred)
  grad = tape.gradient(cost, model.trainable_weights) # 미분값이 구해짐
  optim.apply_gradients(zip(grad, model.trainable_weights)) # update 됨
  print(e, cost)
```

```python
model.evaluate(x_train, y_train, batch_size=128)
```

## 다섯번째 딥러닝 - MNIST 손글씨 이미지 분류

```python
import tensorflow as tf
import pandas as pd

# 데이터를 준비합니다.
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()  # return [(x, y), (x, y)]
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

# 표로 만들기
x_train = x_train.reshape(60000, 784) # 1차원으로 reshape
x_test = x_test.reshape(10000, 784)

# 원핫인코딩
y_train = pd.get_dummies(y_train)
y_test = pd.get_dummies(y_test)

print(x_train.shape, y_train.shape) # 원핫인코딩해서 표(2차원)로 바뀜
print(x_test.shape, y_test.shape)
```

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[784])
H = tf.keras.layers.Dense(64, activation='swish')(X)
Y = tf.keras.layers.Dense(10, activation="softmax")(H)
model = tf.keras.Model(X, Y)
model.compile(loss="categorical_crossentropy", metrics="accuracy")
model.summary()
```

### Batch Normalization

스케일 맞춰서 넣어줬는데, random하게 w 지정되면서, 스케일이 또 깨짐

무조건 써라

```python
X = tf.keras.Input(shape = [784])

H = tf.keras.layers.Dense(196,activation='swish')(X)
H = tf.keras.layers.BatchNormalization()(H)

H = tf.keras.layers.Dropout(0.5)(H)
H = tf.keras.layers.Dense(49,activation='swish')(H)
H = tf.keras.layers.BatchNormalization()(H)

H = tf.keras.layers.Dropout(0.5)(H)
H = tf.keras.layers.Dense(16, activation='swish')(H)
H = tf.keras.layers.BatchNormalization()(H)

H = tf.keras.layers.Dropout(0.3)(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)

model2 = tf.keras.Model(X, Y)
model2.compile(loss="categorical_crossentropy", metrics="accuracy")
model2.summary()
```

### Dropout

최적의 BatchNormalization 위치 Dense 다음

```python
## 최종
X = tf.keras.Input(shape = [784])

H = tf.keras.layers.Dense(196)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.6)(H)
H = tf.keras.layers.Dense(49)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.5)(H)
H = tf.keras.layers.Dense(16)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.3)(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)

model3 = tf.keras.Model(X, Y)
model3.compile(loss="categorical_crossentropy", metrics="accuracy")
model3.summary()
```

0.5, 0.6 을 보통쓰는데,
해당 Dense layer의 차원이 낮은 경우는 0.3정도 쓴다.

```python
# 데이터로 모델을 학습합니다.
model3.fit(x_train, y_train, epochs=100, batch_size=128,
          validation_split=0.2)  # validation_data = (x_val, y_val) 도 있음)
```

loss는 줄어드는데 val의 loss 늘어나면 오버피팅이구나!

```python
# 모델을 평가합니다.
model3.evaluate(x_test, y_test)
```

## MNIST 손글씨 분류 - Flatten

다른 방법론에 대해 말해보자

(1) 데이터 reshape() 안해도 된다 => Flatten()

(2) 숫자로 매칭된 클래스에서 원핫 인코딩 안해도 된다 => sparse_categorical_crossentropy()

```python
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
(x_train, y_train), (x_test, y_test)= tf.keras.datasets.mnist.load_data()

# # 표로 만들기
# x_train = x_train.reshape(60000, 784) # 1차원으로 reshape
# x_test = x_test.reshape(10000, 784)

# # 원핫 인코딩
# y_train = pd.get_dummies(y_train)
# y_test = pd.get_dummies(y_test)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### sparse_categorical_crossentropy

숫자로 매칭이 된 클래스인 경우, 원핫 인코딩 안해줘도 됨 => sparse_categorical_crossentropy

```python
# 모델 준비
X = tf.keras.Input(shape=[28,28])
H = tf.keras.layers.Flatten()(X)
H = tf.keras.layers.Dense(64, activation="swish")(H)
Y = tf.keras.layers.Dense(10, activation="softmax")(H)
model = tf.keras.Model(X, Y)
model.compile(loss="sparse_categorical_crossentropy", metrics="accuracy")
model.summary()
```

```python
# 모델 학습
result = model.fit(x_train, y_train, epochs=100, validation_split=0.1, batch_size=128)
result = model.fit(x_train, y_train, epochs=10, validation_split=0.1, batch_size=128)
```

### 문자열 하이퍼파라미터, 모두 클래스의 속성으로 대체 가능

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[28, 28])
H = tf.keras.layers.Flatten()(X)
H = tf.keras.layers.Dense(64, activation=tf.keras.activations.swish)(H)
Y = tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax)(H)
model = tf.keras.Model(X, Y)
model.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
              metrics=tf.keras.metrics.sparse_categorical_accuracy)
model.summary()
```

## 실습. CIFAR10 학습하기
- 정확도: 50% 넘겨보기


```python
import tensorflow as tf

# cifar10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

x_train = x_train / 255.
x_test = x_test / 255.
```

```python
import matplotlib.pyplot as plt

print("정답:", y_train[0])
plt.imshow(x_train[0])
plt.show()
```

```python
X = tf.keras.Input(shape=[32, 32, 3])
H = tf.keras.layers.Flatten()(X)

H = tf.keras.layers.Dropout(0.7)(H)
H = tf.keras.layers.Dense(128)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.6)(H)
H = tf.keras.layers.Dense(128)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.6)(H)
H = tf.keras.layers.Dense(128)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.6)(H)
H = tf.keras.layers.Dense(128)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dropout(0.6)(H)
Y = tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax)(H)
model = tf.keras.Model(X, Y)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              # loss=tf.keras.losses.sparse_categorical_crossentropy,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              # metrics=tf.keras.metrics.sparse_categorical_accuracy,
              metrics=tf.keras.metrics.SparseCategoricalAccuracy())
model.summary()
```

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[32, 32, 3])
H = tf.keras.layers.Flatten()(X)

H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(128, activation="swish")(H)
H = tf.keras.layers.Dropout(0.5)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(32, activation="swish")(H)
H = tf.keras.layers.Dropout(0.5)(H)
Y = tf.keras.layers.Dense(10, activation="softmax")(H)
model = tf.keras.Model(X, Y)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=tf.keras.metrics.SparseCategoricalAccuracy())
```

```python
# 모델을 준비합니다.
# X = tf.keras.Input(shape=[28, 28])
# H = tf.keras.layers.Flatten()(X)
# H = tf.keras.layers.Dense(64, activation=tf.keras.activations.swish)(H)
# H = tf.keras.layers.BatchNormalization(H)
# H = tf.keras.layers.Dense(20, activation=tf.keras.activations.swish)(H)
# H = tf.keras.layers.BatchNormalization(H)
# Y = tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax)(H)
# model = tf.keras.Model(X, Y)
# model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
#               # loss=tf.keras.losses.sparse_categorical_crossentropy,
#               loss=tf.keras.losses.SparseCategoricalCrossentropy(),
#               # metrics=tf.keras.metrics.sparse_categorical_accuracy,
#               metrics=tf.keras.metrics.SparseCategoricalAccuracy())
# model.summary()


# loss = tf.keras.losses.MeanSquaredError()
# optim = tf.keras.optimizers.Adam(learning_rate=0.001)

# for e in range(1000):
#   with tf.GradientTape() as tape: # 미분을 할 수 있는 기능
#     pred = model(np.array(x_train), training=True)
#     cost = loss(np.array(y_train), pred)
#   grad = tape.gradient(cost, model.trainable_weights # 미분값이 구해짐
#   optim.apply_gradients(zip(grad, model.trainable_weights) # 그 미분값을 가지고, weigh들을 update
#   print(e, cost)


class MyFashionMNISTModel(tf.keras.Model):
    def __init__(self, **kwargs):
        super(MyFashionMNISTModel, self).__init__(**kwargs)
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(64, activation="swish")
        self.bn1 = tf.keras.layers.BatchNormalization()
        self.dense2 = tf.keras.layers.Dense(20, activation="swish")
        self.bn2 = tf.keras.layers.BatchNormalization()
        self.dense3 = tf.keras.layers.Dense(10, activation="softmax")

        # self.skip_connection_layers = []
        # for i in range(16):
        #     dropout = tf.keras.layers.Dropout(0.5)
        #     dense = tf.keras.layers.Dense(512)
        #     bn = tf.keras.layers.BatchNormalization()
        #     activation = tf.keras.layers.Activation('swish')
        #     self.skip_connection_layers.append([dropout, dense, bn, activation])

    def call(self, X):
        H = self.flatten(X)
        H = self.dense1(H)
        H = self.bn1(H)
        H = self.dense2(H)
        H = self.bn2(H)
        Y = self.dense3(H)

        # for drop, dense, bn, activation in self.skip_connection_layers:
        #     H1 = drop(H)
        #     H1 = dense(H1)
        #     H1 = bn(H1)
        #     H = activation(H1)

        return Y

    def train_step(self, batch): # batch 형태로 input
      x_batch, y_batch = batch
      # print(x_batch)

      with tf.GradientTape() as tape:
        y_pred = self(x_batch, training=True)
        loss = self.compiled_loss(y_batch, y_pred)

      grad = tape.gradient(loss, self.trainable_weights)
      self.optimizer.apply_gradients(zip(grad, self.trainable_weights))

      self.compiled_metrics.update_state(y_batch, y_pred) # 누적된 점
      return {m.name: m.result() for m in self.metrics}

    # evaluate 할 때 호출
    def test_step(self, batch):
      x_batch, y_batch = batch
      y_pred = self(x_batch)

      self.compiled_loss(y_batch, y_pred)
      self.compiled_metrics.update_state(y_batch, y_pred)
      return {m.name: m.result() for m in self.metrics}

model = MyFashionMNISTModel()
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics="accuracy")
model.build(input_shape=[None, 28, 28]) # None, 데이터의 개수
model.summary()
```

- 함수 + packing/unpacking

```python
def add(a, b, c):
  return a+b+c

## unpacking
d = [1,2,3]
print(add(*d))

dic = {'a':1, 'b':2, 'c':3}
print(add(**dic)) # 변수명과 키가 일치하는 값이 들어감 (순서대로가 아님)


## packing
def add(*a):
  print(a)
  return sum(a)

print(add(1, 2, 3, 4))

def add(**a):
  print(a)
  return sum(a.values())

print(add(a=1, b=2, c=3))
```

```python
# 데이터 준비
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# normalize
x_train = x_train / 255
x_test = x_test / 255

# 텐서플로우의 원핫인코딩 함수
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# 모델 생성
X = tf.keras.Input(shape=[32, 32, 3])

H = tf.keras.layers.Conv2D(6, 5)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)
# H = tf.keras.layers.MaxPool2D()(H)
H = tf.keras.layers.Conv2D(6, 5, strides=2, padding="same")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)

H = tf.keras.layers.Conv2D(16, 5)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)
# H = tf.keras.layers.MaxPool2D()(H)
H = tf.keras.layers.Conv2D(16, 5, strides=2, padding="same")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)

H = tf.keras.layers.Flatten()(H)

H = tf.keras.layers.Dropout(0.6)(H)
H = tf.keras.layers.Dense(120)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)

H = tf.keras.layers.Dense(84)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("swish")(H)

Y = tf.keras.layers.Dense(10, activation="softmax")(H)

model = tf.keras.Model(X, Y)
model.compile(loss="categorical_crossentropy", metrics="accuracy")
model.summary()
```

```python
# 모델 학습
es = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
model.fit(x_train, y_train, epochs=100000, batch_size=128, validation_split=0.1, callbacks=[es])
```

## Data Augmentation
- 256x256 이미지 10,000장을 학습용 이미지로 제공
- 224x224로 crop 하여 1장의 이미지에서 1024 이미지를 추출
- 좌우반전을 더해 총 2048만 장의 이미지로 학습을 하게 된다.

```python
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=False,
    rotation_range=0.1,
)
train_ds = datagen.flow(x_train[:40000], y_train[:40000], batch_size=128)
valid_ds = tf.data.Dataset.from_tensor_slices((x_train[40000:], y_train[40000:])).batch(128)
```

```python
x, y = next(iter(train_ds))
print(x.shape, y.shape)

import matplotlib.pyplot as plt
plt.imshow(x[0])
plt.show()
```

```python
# 모델 학습
es = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
model.fit(train_ds, validation_data=valid_ds, epochs=100000, callbacks=[es])
```

```python
model.evaluate(x_test, y_test)
```
