#### **<span style="color:orange">[2-1] air_21, air_22 의 '측정일시'를 활용하여 'time' 변수 생성</span>**

* air_21, air_22  각각 '측정일시'를 활용하여 'time'변수 생성
    * 참고: 미세먼지 데이터는 1시-24시, 날씨 데이터는 0시-23시로 구성되어 있습니다. [2-3]에서 미세먼지와 날씨 데이터를 time 기준으로 합치려면 기준이 동일해야 합니다. 미세먼지 데이터에서 time 변수 생성 시 이를 미리 고려(예: air_21['측정일시'] -1)하세요.
* time 변수를 pd.to_datetime으로 데이터 타입 변경
    * 참고: format = '%Y%m%d%H'

```python
air_21.head()
```

```python
air_21['time'] = pd.to_datetime(air_21.측정일시.astype(str).str[:-2], format='%Y%m%d')

air_21['hours'] = air_21.측정일시.astype(str).str[-2:].astype(int)
air_21.loc[air_21['hours']==24, 'time'] = air_21['time'] + pd.Timedelta(days=1)
air_21.loc[air_21['hours']==24, 'hours'] = 0

air_21['time'] = air_21['time'] + pd.to_timedelta(air_21['hours'], unit='h')

air_21
```

```python
air_22['time'] = pd.to_datetime(air_22.측정일시.astype(str).str[:-2], format='%Y%m%d')

air_22['hours'] = air_22.측정일시.astype(str).str[-2:].astype(int)
air_22.loc[air_22['hours']==24, 'time'] = air_22['time'] + pd.Timedelta(days=1)
air_22.loc[air_22['hours']==24, 'hours'] = 0

air_22['time'] = air_22['time'] + pd.to_timedelta(air_22['hours'], unit='h')

air_22
```
