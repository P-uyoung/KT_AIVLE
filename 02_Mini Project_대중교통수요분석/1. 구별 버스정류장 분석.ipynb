```python
# bus_station과 only_seoul 데이터를 merge 해봅니다.
bus_station = pd.merge(user1, bus, how='inner', on='버스정류장ARS번호')
```

#### [1.2.7] 구 별로 버스 정류장의 개수 확인

```python
# 구 별로 버스 정류장의 개수를 확인해 봅니다.
# 구 별 버스정류장 개수로 구성된 데이터프레임을 seoul_bus_station_ARS에 저장합니다.(groupby 함수에서 as_index 옵션을 사용해보세요)
seoul_bus_station_ARS = pd.DataFrame(bus_station.groupby('자치구', as_index=True)['버스정류장ARS번호'].nunique())
seoul_bus_station_ARS.columns = ['버스정류장개수']
seoul_bus_station_ARS.head()
```

#### [1.2.8] 구 별로 버스 노선이 몇 개 지나가는 지 확인
* 구 별로 버스 노선의 개수 확인

```python
# 구 별로 버스 노선 개수를 확인해 봅니다.
# 구 별 버스 노선 개수로 구성된 데이터프레임을 seoul_bus_station_line에 저장합니다.
seoul_bus_station_line = pd.DataFrame(bus_station.groupby('자치구', as_index=True)['노선번호'].nunique())
seoul_bus_station_line.columns = ['버스노선개수']
seoul_bus_station_line.head()
```

#### [1.2.9] 구 별로 승차 총 승객수, 하차 총 승객수 확인

```python
# 구 별 승차총승객수, 하차총승객수를 확인해봅니다.
# 구 별 승차총승객수, 하차총승객수로 구성된 데이터프레임을 seoul_bus_station_sum에 저장합니다.
seoul_bus_station_sum = pd.DataFrame(bus_station.groupby('자치구', as_index=True)['승차총승객수'].sum())
seoul_bus_station_sum['하차총승객수'] =  pd.DataFrame(bus_station.groupby('자치구', as_index=True)['하차총승객수'].sum()).values
seoul_bus_station_sum
```

#### [1.2.10] 구 별로 승차 평균 승객수, 하차 평균 승객수 확인

```python
# 구 별 승차 승객수, 하차 승객수의 평균을 확인해봅니다.
# 구 별 승차 승객수, 하차 승객수의 평균으로 구성된 데이터프레임을 seoul_bus_station_mean에 저장합니다.
# 컬럼명은 승차평균승객수, 하차평균승객수로 바꾸세요.
seoul_bus_station_mean = pd.DataFrame(bus_station.groupby('자치구', as_index=True)['승차총승객수'].mean())
seoul_bus_station_mean['하차총승객수'] =  pd.DataFrame(bus_station.groupby('자치구', as_index=True)['하차총승객수'].mean()).values
seoul_bus_station_mean.columns = ['승차승객수평균', '하차승객수평균']
seoul_bus_station_mean
```

#### [1.3.1] 데이터 분포 알아보기
* 다양한 변수를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
# 아래에 코드를 작성하고 결과를 확인합니다.

seoul_bus_station.reset_index(inplace=True)

sorted_seoul_bus_station = seoul_bus_station.sort_values(by='버스정류장개수', ascending=False)

sns.barplot(x='자치구', y='버스정류장개수', data=sorted_seoul_bus_station)
plt.xticks(rotation=70)
plt.show()
```

```python
sorted_seoul_bus_station = seoul_bus_station.sort_values(by='버스노선개수', ascending=False)

sns.barplot(x='자치구', y='버스노선개수', data=sorted_seoul_bus_station)
plt.xticks(rotation=70)
plt.show()
```

```python
sorted_seoul_bus_station = seoul_bus_station.sort_values(by='버스노선개수', ascending=False)
melted_seoul_bus_station = seoul_bus_station.melt(id_vars='자치구', 
                                                  value_vars=['승차총승객수', '하차총승객수'],
                                                  var_name='승하차',
                                                  value_name='승하차 승객')
sns.barplot(x='자치구', y='승하차 승객', hue='승하차', data=melted_seoul_bus_station)
plt.xticks(rotation=70)
plt.show()
```

```python
seoul_bus_station['승하차승객차이'] = seoul_bus_station['승차총승객수'] - seoul_bus_station['하차총승객수']
sorted_seoul_bus_station = seoul_bus_station.sort_values(by='승하차승객차이', ascending=False)

sns.barplot(x='자치구', y='승하차승객차이', data=sorted_seoul_bus_station)
plt.xticks(rotation=70)
plt.show()
```

위 차트를 통해 알게된 사실을 정리해봅시다.
1. 구별 정류장 개수 : 성복구 / 중구
2. 구별 노선수 : 서대문구 / 강동구
3. 구별 승하차 고객수 : 강남구가 하차 인원과 승차인원의 차이가 크다.
