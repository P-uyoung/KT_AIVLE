# 1. 데이터 전처리

#### [실습문제1] 열 추가
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['year'], df_total['month'], df_total['day'], df_total['weekday'] = df_total['기준일ID'].dt.year, df_total['기준일ID'].dt.month, df_total['기준일ID'].dt.day, df_total['기준일ID'].dt.weekday
df_total.head(3)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_test['year'], df_test['month'], df_test['day'], df_test['weekday'] = df_test['기준일ID'].dt.year, df_test['기준일ID'].dt.month, df_test['기준일ID'].dt.day, df_test['기준일ID'].dt.weekday
df_test.info()
```

#### [실습문제2] 불필요한 열 제거 및 열 이름 변경
* df_total과 df_test에서 '총생활인구수'외에 불필요한 열을 제거 하도록 하겠습니다.

```python
train = df_total
test = df_test
```

```python
train['hour'] = train['시간대구분']
test['hour'] = test['시간대구분']
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

#### shift()로 전후 평균

```python
train['mean3'] = round(train['총생활인구수'].rolling(3, min_periods=1).mean(), 1)
train.head()
```

```python
test['mean3'] = round(test['총생활인구수'].rolling(3, min_periods=1).mean(), 1)
test.head()
```

#### target 값 설정

요일도 중요할 것 같아서 일주일 뒤 총생활인구수로 target을 설정했는데, 0.6 성능 나옴.

하루 뒤 총생활인구수로 target을 설정하겠음.

```python
# target : 일주일 뒤 총생활인구수
train = pd.concat([train, test.iloc[:24]], axis=0)  # 행으로 합치기
train['target'] = train['총생활인구수'].shift(-24)
train.info()
```

```python
train.dropna(inplace=True) # null 값을 포함하는 행 삭제
train
```

```python
test['target'] = test['총생활인구수'].shift(-24)
test.dropna(inplace=True) # null 값을 포함하는 행 삭제
test
```

#### [실습문제5] train, test 데이터 나누기
* df_total과 df_test를 각각 train_x, train_y, test_x, test_y 로 나누어 주세요.
* 인덱스를 리셋 해주세요.

```python
train.info()
```

```python
train.drop(['year', 'day'], axis=1, inplace=True)
test.drop(['year', 'day'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoder 초기화 (sparse_output 매개변수 사용)
encoder = OneHotEncoder(sparse_output=False)

# ['year', 'month', 'day', 'hour', 'weekday'] 컬럼들을 one-hot encoding
encoded_cols = encoder.fit_transform(train[['month', 'hour', 'weekday']])
encoded_train = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(['month', 'hour', 'weekday']))

# 원래의 데이터프레임에서 해당 컬럼들을 제거하고 one-hot encoded 데이터를 추가
train = train.drop(columns=['month', 'hour', 'weekday'])
train = pd.concat([train, encoded_train], axis=1)
```

```python
encoded_cols = encoder.transform(test[['month', 'hour', 'weekday']])
encoded_test = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(['month', 'hour', 'weekday']))

# 원래의 데이터프레임에서 해당 컬럼들을 제거하고 one-hot encoded 데이터를 추가
test = test.drop(columns=['month', 'hour', 'weekday'])
test = pd.concat([test, encoded_test], axis=1)
```

```python
train_x = train.drop('target', axis=1)
train_y = train['target']
test_x = test.drop('target', axis=1)
test_y = test['target']
```
