```python
import cv2
import torch

# YOLOv5 라이브러리에서 모델 로드
model = torch.hub.load('ultralytics/yolov5', 'custom', path='./best.pt')  # 'custom'은 사용자 정의 모델을 의미합니다.

cap = cv2.VideoCapture(2)

while True:
    ret, img = cap.read()
    if not ret:
        break

    # 모델 예측
    results = model(img, size=640)  # size는 입력 이미지의 크기입니다. 필요에 따라 조정할 수 있습니다.
    # rendered_img = results.render()[0]  # 예측 결과를 이미지에 렌더링
    img = result[0].plot()
    cv2.imshow('Bill Detection', img)
    # cv2.imshow('Bill Detection', rendered_img)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

yolov8일때,

```python
from ultralytics import YOLO
import cv2
cap = cv2.VideoCapture(2)

model = YOLO(model='./best.pt', task='detect')
while True:
  ret, img = cap.read()
  result = model.predict(source=img, conf=0.25, iou=0.7, save=False,
                         line_width=2)
  img = result[0].plot()

  cv2.imshow('Bill Detection', img)

  if cv2.waitKey(10) == ord('q'):
    break
cap.release()
cv2.destroyAllWindows()
```
